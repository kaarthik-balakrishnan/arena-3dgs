````python
#@title Upload Images
from google.colab import files
import os
import zipfile

# Create input directory
!mkdir -p /content/input_images

# Upload images
print("Please upload a ZIP file containing your images")
uploaded = files.upload()

# Extract if it's a zip
for fname in uploaded.keys():
    if fname.endswith('.zip'):
        with zipfile.ZipFile(fname, 'r') as zip_ref:
            zip_ref.extractall('/content/input_images')

# Count images
img_count = len([f for f in os.listdir('/content/input_images') if f.endswith(('.jpg', '.jpeg', '.png'))])
print(f"Uploaded {img_count} images")
````

````python
#@title Run COLMAP for Camera Poses
import subprocess

input_path = "/content/input_images"
output_path = "/content/colmap_output"

os.makedirs(output_path, exist_ok=True)

# Run COLMAP
colmap_cmds = [
    f"colmap feature_extractor --database_path {output_path}/database.db --image_path {input_path}",
    f"colmap exhaustive_matcher --database_path {output_path}/database.db",
    f"colmap mapper --database_path {output_path}/database.db --image_path {input_path} --output_path {output_path}"
]

for cmd in colmap_cmds:
    print(f"Running: {cmd.split()[0]}...")
    subprocess.run(cmd.split(), capture_output=True)

# Check output
!ls {output_path}/0/
````
